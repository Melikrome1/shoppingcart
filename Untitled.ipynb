```python

shopping_cart = {}

def add_to_cart():
    item = input("Enter the item name: ")
    quantity = int(input("Enter the quantity: "))
    price = float(input("Enter the price per unit: "))
    # Check if the item already exists in the shopping cart
    if item in shopping_cart:
        shopping_cart[item]["quantity"] += quantity
        shopping_cart[item]["price"] += price * quantity
    else:
        shopping_cart[item] = {"quantity": quantity, "price": price * quantity}
    print(f"{quantity} {item}(s) added to the cart.")

# Define a function to remove an item from the shopping cart
def remove_from_cart():
    item = input("Enter the item name: ")
    # Check if the item exists in the shopping cart
    if item in shopping_cart:
        quantity = int(input("Enter the quantity to remove: "))
        # Check if the requested quantity is less than or equal to the quantity in the cart
        if quantity <= shopping_cart[item]["quantity"]:
            shopping_cart[item]["quantity"] -= quantity
            shopping_cart[item]["price"] -= shopping_cart[item]["price"] / shopping_cart[item]["quantity"] * quantity
            # Remove the item from the cart if the quantity becomes zero
            if shopping_cart[item]["quantity"] == 0:
                del shopping_cart[item]
            print(f"{quantity} {item}(s) removed from the cart.")
        else:
            print("The requested quantity is not available in the cart.")
    else:
        print("The item is not in the cart.")


def view_cart():
    if shopping_cart:
        print("Items in the cart:")
        for item in shopping_cart:
            print(f"{item}: {shopping_cart[item]['quantity']} x ${shopping_cart[item]['price']/shopping_cart[item]['quantity']:.2f} = ${shopping_cart[item]['price']:.2f}")
        print(f"Total price: ${sum(shopping_cart[item]['price'] for item in shopping_cart):.2f}")
    else:
        print("The cart is empty.")

# Define a function to clear the shopping cart
def clear_cart():
    shopping_cart.clear()
    print("The cart is cleared.")

# Define a function to start the shopping cart program
def start_shopping_cart():
    print("Welcome to the shopping cart program!")
    while True:
        print("\nWhat would you like to do?")
        print("1. Add item to cart")
        print("2. Remove item from cart")
        print("3. View cart")
        print("4. Clear cart")
        print("5. Quit/Check out")
        choice = input("Enter your choice (1-5): ")
        if choice == "1":
            add_to_cart()
        elif choice == "2":
            remove_from_cart()
        elif choice == "3":
            view_cart()
        elif choice == "4":
            clear_cart()
        elif choice == "5":
            print("Thank you for shopping with us!")
            view_cart()
            break
        else:
            print("Invalid choice. Please try again.")

# Call the start_shopping_cart function to start the program
start_shopping_cart()
```
